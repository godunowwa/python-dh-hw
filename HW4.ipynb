```python
import nltk
from nltk import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string
nltk.download('punkt')
nltk.download('wordnet')
f = open('Harry.txt', 'r')
raw_data = f.read()
f.close()
raw_data = [w.strip(string.punctuation).lower() for w in word_tokenize(raw_data) if w not in string.punctuation and w != '«' and w != '»']
data = [WordNetLemmatizer().lemmatize(w, pos='v') for w in raw_data]
f = open('HarryL.txt', 'w')
f.write(' '.join(data))
f.close()
```

```python
from collections import Counter
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
words = Counter([i for i in data])
k = len(data)
j = len(words)
x_points = [x_point+1 for x_point in range(j)]
y_points = sorted([y_point/k for y_point in words.values()], reverse=True)
plt.plot(x_points, y_points)
ax.set_xlabel('Ранг')
ax.set_ylabel('Относительная частота')
plt.show()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
import numpy
filtered_data = [w for w in data if w not in stopwords.words('english')]
words = Counter([i for i in filtered_data])
most_common_wors = words.most_common(30)
fig, ax = plt.subplots()
plt.bar([x for x in range(30)], [x[1] for x in most_common_wors])
ax.set_xticks(numpy.arange(len(most_common_wors)))
ax.set_xticklabels([w[0] for w in most_common_wors])
ax.set_xlabel('Слова')
ax.set_ylabel('Абсолютная частота')
plt.show()
```
